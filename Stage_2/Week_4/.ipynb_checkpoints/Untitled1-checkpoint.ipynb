```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('clean_googleplaystore_user_reviews.csv')


df_n = pd.read_excel('C:/Users/DBSOM-L490-24/Documents/DEVELHOPE/PYTHON/data-5---team-4/Stage_2/Data/n.xlsx')
df_p = pd.read_excel('C:/Users/DBSOM-L490-24/Documents/DEVELHOPE/PYTHON/data-5---team-4/Stage_2/Data/p.xlsx')

print(f'the shape of n is {df_n.shape}')
print(f'the shape of p is {df_p.shape}')
print(f'the shape of df_review is {df.shape}')
```

```python
import re
import string


def sent_cat(data):
    if data > 0:
        return "positive"
    elif data < 0:
        return "negative"
    else:
        return "neutral"
    
def sen_p(data):
    sent_scor = 0
    for positive in df_p.itertuples():
        if positive[0] in data:
            sent_scor += 1
    for negative in df_n.itertuples():
        if positive[0] in data:
            sent_scor -= 1
    return sent_scor      
        
def filt_str(row):
    return re.sub('['+re.escape(string.punctuation)+']', '',row)

df['Translated_Reviews']= df['Translated_Review'].apply(lambda row : filt_str(row).lower())

#df['Sentimental_Score']=df['Translated_Review'].apply(lambda row : sen_p(list(row) ))

#df_review['Sentimental_Cat']=df_review['Translated_Review'].apply(lambda row : sent_cat(row))

df
```
